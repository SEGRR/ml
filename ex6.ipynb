```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#reading the file
dataset=[]
with open('/content/Market_Basket_Optimisation.csv') as file:
  reader=csv.reader(file,delimiter=',')
  for row in reader:
    dataset += [row]
```

```python
dataset[0:10]
```

```python
len(dataset) #checking the length of the dataset
```

```python
te =TransactionEncoder()
x=te.fit_transform(dataset)
```

```python
x
```

```python
df=pd.DataFrame(x,columns=te.columns_)
```

```python
df.head()
```

```python
#1.frequent itets dataset
freq_itemset = apriori(df,min_support=0.01,use_colnames=True)
#candidate itemset
```

```python
freq_itemset
```

```python
#find the rules
rules=association_rules(freq_itemset,metric='confidence',min_threshold=0.25)
```

```python
rules=rules[['antecedents',"consequents","support","confidence"]]
print(rules)
```
