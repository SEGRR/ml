```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/TE Labs/ML/datasets/SMSSpamCollection" , sep="\t" , )
```

```python
dataset.head()
```

```python
dataset.groupby("category").describe()
```

```python
dataset.isnull().sum()
```

```python
dataset['spam'] = dataset['category'].apply(lambda x: 1 if x == "spam" else 0)
```

```python
dataset.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(dataset.message , dataset.spam)
```

```python
x_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x_train = cv.fit_transform(x_train.values)
```

```python
print(cv.get_feature_names_out())
```

```python
type(x_train)
```

```python
x_train_array = x_train.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train , y_train)
```

```python
x_test_count = cv.transform(x_test)
y_pred = model.predict(x_test_count)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test , y_pred))
print("Calssification Report ")
print(classification_report(y_test ,y_pred))
```

```python
emails =['How are you brother?' ,'Get Flat on Your body outfits ']
emails_count = cv.transform(emails)
model.predict(emails_count)
```

```python
def classify(message):
  vector = cv.transform([message])
  pre = model.predict(vector)
  return "SPAM" if pre[0] == 1 else "NOT SPAM"
```
